```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Excels/sonar.csv',header=None)
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df[60].value_counts()
```

```python
X = df.drop(60,axis='columns')
y = df[60]
```

```python
y = pd.get_dummies(y,drop_first=True)
y.head()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.25,random_state=5)
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'), # input 
    keras.layers.Dense(30,activation='relu'), # hidden layer
    keras.layers.Dense(15,activation='relu'), # hidden layer
    keras.layers.Dense(1,activation='sigmoid') # 1 because binary classifcation
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100,batch_size=8) # batch_size is mini batch which contain 8 samples only
```

```python
model.evaluate(X_test,y_test) # becuase i overfit in my model.fit , bad score
```

```python
y_pred = model.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
y_pred[:5]
```

```python
y_test[:5]
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_test,y_pred))
```

# Dropout Regulartziation 

```python
model = keras.Sequential([
    keras.layers.Dense(60,input_shape=(60,),activation='relu'), # input 
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'), # hidden layer
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'), # hidden layer
    keras.layers.Dropout(0.5),    
    keras.layers.Dense(1,activation='sigmoid') # 1 because binary classifcation
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100,batch_size=8) # batch_size is mini batch which contain 8 samples only
```

```python
model.evaluate(X_test,y_test) # lower trainning set score , but more on tes
```

```python
y_pred = model.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
print(classification_report(y_test,y_pred))
```
